**Project Goal:** Analyze sales, profit, and customer satisfaction across product categories and regions for NovaMart, a fictional online electronics store.


## 1️⃣ Import Libraries & Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (replace with actual file path if available)
data = pd.read_csv('novamart_sales.csv')
data.head()
```

## 2️⃣ Data Overview

```python
data.info()
data.describe()
```

## 3️⃣ Sales Performance by Category

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Sales', data=data, estimator=sum, ci=None)
plt.title('Total Sales by Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 4️⃣ Regional Profit Distribution

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Region', y='Profit', data=data)
plt.title('Profit Distribution by Region')
plt.tight_layout()
plt.show()
```

## 5️⃣ Insights Summary
- Highest sales category: Electronics
- Region with lowest profit margin: East
- Recommendation: Optimize pricing in low-profit regions and promote top-selling categories.
